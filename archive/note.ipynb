### 4. Risks, Constraints & Auditor Analysis (Fact-Check)

As part of our governance and security review, please note the following architectural risks and industry best-practice gaps identified during the meeting:

* **Risk 1: Manual Compliance Gap (High Risk)**
* *Issue*: Relying on business users to manually run the Solution Checker and honestly upload the latest code to SharePoint is prone to human error and bypassing. There is no guarantee the `.zip` on SharePoint matches the version running in production.


* **Risk 2: SharePoint as a Code Repository (Process Gap)**
* *Issue*: SharePoint lacks Git-based version control features (Diff, Branching) and cannot natively trigger CI/CD pipelines. Furthermore, SAST tools (like Snyk) cannot effectively scan packed `.zip` files without a backend unpacking script (`pac solution unpack`).


* **Risk 3: Licensing Misconception vs. Best Practices**
* *Fact-Check*: It was stated that moving code to Azure DevOps requires all users to have Premium/DevOps licenses. **This is not entirely accurate.** * *Best Practice*: We can utilize a single Service Principal (Service Account) with Environment Maker/Admin privileges. A backend pipeline can automatically pull (Export & Unpack) code from all environments to Azure DevOps on a schedule. This provides InfoSec with automated scanning and version control *without* requiring business users to learn Git, manually upload files, or incur individual Premium licensing costs. **(Highly recommend exploring AI-005/AI-006 to validate this).**


* **Risk 4: DLP Configuration & Visibility**
* *Issue*: InfoSec has had zero visibility into Power Platform adoption since last November, and misconfigured DLP buckets were noted.
* *Mitigation*: Expedite AI-009 (CoE Starter Kit) to deploy the *Power BI Dashboard for Adoption* and *Environment/DLP Monitoring* immediately.
